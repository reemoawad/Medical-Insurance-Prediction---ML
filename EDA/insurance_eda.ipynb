```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
# Set visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# ============================================================================
# 1. DATA LOADING
# ============================================================================

print("="*80)
print("MEDICAL INSURANCE COST PREDICTION - EDA")
print("="*80)

# Load the dataset
df = pd.read_csv("../insurance.csv")

print("\n📊 Dataset loaded successfully!")
print(f"Shape: {df.shape[0]} rows, {df.shape[1]} columns")
```

```python
# ============================================================================
# 2. INITIAL DATA INSPECTION
# ============================================================================

print("\n" + "="*80)
print("2. INITIAL DATA INSPECTION")
print("="*80)

print("\n🔍 First 5 rows:")
print(df.head())

print("\n📋 Dataset Info:")
print(df.info())

print("\n📊 Statistical Summary:")
print(df.describe())

print("\n🔢 Data Types:")
print(df.dtypes)
```

```python
# ============================================================================
# 3. MISSING VALUES & DATA QUALITY
# ============================================================================

print("\n" + "="*80)
print("3. DATA QUALITY CHECK")
print("="*80)

print("\n❓ Missing Values:")
missing = df.isnull().sum()
print(missing)
print(f"\nTotal missing values: {missing.sum()}")

print("\n🔁 Duplicate Rows:")
duplicates = df.duplicated().sum()
print(f"Number of duplicates: {duplicates}")
```

```python
# ============================================================================
# 4. FEATURE ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("4. FEATURE ANALYSIS")
print("="*80)

# Categorical features
print("\n📌 Categorical Features Distribution:")
categorical_cols = ['sex', 'smoker', 'region']

for col in categorical_cols:
    print(f"\n{col.upper()}:")
    print(df[col].value_counts())
    print(f"Percentage:\n{df[col].value_counts(normalize=True)*100}")

# Numerical features
print("\n📊 Numerical Features Summary:")
numerical_cols = ['age', 'bmi', 'children']
print(df[numerical_cols].describe())

# Target variable
print("\n💰 TARGET VARIABLE (charges):")
print(f"Mean: ${df['charges'].mean():.2f}")
print(f"Median: ${df['charges'].median():.2f}")
print(f"Std Dev: ${df['charges'].std():.2f}")
print(f"Min: ${df['charges'].min():.2f}")
print(f"Max: ${df['charges'].max():.2f}")
```

```python
# ============================================================================
# 5. VISUALIZATIONS
# ============================================================================

print("\n" + "="*80)
print("5. CREATING VISUALIZATIONS")
print("="*80)

# Distribution of target variable
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].hist(df['charges'], bins=50, edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Insurance Charges ($)')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Distribution of Insurance Charges')
axes[0].axvline(df['charges'].mean(), color='red', linestyle='--', label=f'Mean: ${df["charges"].mean():.2f}')
axes[0].axvline(df['charges'].median(), color='green', linestyle='--', label=f'Median: ${df["charges"].median():.2f}')
axes[0].legend()

axes[1].hist(np.log(df['charges']), bins=50, edgecolor='black', alpha=0.7, color='orange')
axes[1].set_xlabel('Log(Insurance Charges)')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Distribution of Log-Transformed Charges')

plt.tight_layout()
plt.savefig('charges_distribution.png', dpi=300, bbox_inches='tight')
print("✅ Saved: charges_distribution.png")
plt.show()
```

```python
# Numerical features distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].hist(df['age'], bins=30, edgecolor='black', alpha=0.7, color='skyblue')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Age Distribution')

axes[1].hist(df['bmi'], bins=30, edgecolor='black', alpha=0.7, color='lightgreen')
axes[1].set_xlabel('BMI')
axes[1].set_ylabel('Frequency')
axes[1].set_title('BMI Distribution')

axes[2].bar(df['children'].value_counts().sort_index().index, 
            df['children'].value_counts().sort_index().values, 
            edgecolor='black', alpha=0.7, color='lightcoral')
axes[2].set_xlabel('Number of Children')
axes[2].set_ylabel('Frequency')
axes[2].set_title('Children Distribution')

plt.tight_layout()
plt.savefig('numerical_distributions.png', dpi=300, bbox_inches='tight')
print("✅ Saved: numerical_distributions.png")
plt.show()
```

```python
# Categorical features vs charges
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.boxplot(data=df, x='sex', y='charges', ax=axes[0], palette='Set2')
axes[0].set_title('Charges by Sex')
axes[0].set_ylabel('Insurance Charges ($)')

sns.boxplot(data=df, x='smoker', y='charges', ax=axes[1], palette='Set1')
axes[1].set_title('Charges by Smoking Status')
axes[1].set_ylabel('Insurance Charges ($)')

sns.boxplot(data=df, x='region', y='charges', ax=axes[2], palette='Set3')
axes[2].set_title('Charges by Region')
axes[2].set_ylabel('Insurance Charges ($)')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('categorical_vs_charges.png', dpi=300, bbox_inches='tight')
print("✅ Saved: categorical_vs_charges.png")
plt.show()
```

```python
# Numerical features vs charges
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].scatter(df['age'], df['charges'], alpha=0.5, color='blue')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Insurance Charges ($)')
axes[0].set_title('Age vs Charges')

axes[1].scatter(df['bmi'], df['charges'], alpha=0.5, color='green')
axes[1].set_xlabel('BMI')
axes[1].set_ylabel('Insurance Charges ($)')
axes[1].set_title('BMI vs Charges')

axes[2].scatter(df['children'], df['charges'], alpha=0.5, color='red')
axes[2].set_xlabel('Number of Children')
axes[2].set_ylabel('Insurance Charges ($)')
axes[2].set_title('Children vs Charges')

plt.tight_layout()
plt.savefig('numerical_vs_charges.png', dpi=300, bbox_inches='tight')
print("✅ Saved: numerical_vs_charges.png")
plt.show()
```

```python
# Smoker effect on age-charges relationship
fig, ax = plt.subplots(figsize=(12, 6))
for smoker_status in df['smoker'].unique():
    subset = df[df['smoker'] == smoker_status]
    ax.scatter(subset['age'], subset['charges'], 
               label=f'Smoker: {smoker_status}', alpha=0.6, s=50)
ax.set_xlabel('Age')
ax.set_ylabel('Insurance Charges ($)')
ax.set_title('Age vs Charges (colored by Smoking Status)')
ax.legend()
plt.savefig('age_charges_smoker.png', dpi=300, bbox_inches='tight')
print("✅ Saved: age_charges_smoker.png")
plt.show()
```

```python
# BMI categories and smoking
df['bmi_category'] = pd.cut(df['bmi'], 
                              bins=[0, 18.5, 25, 30, 100], 
                              labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(data=df, x='bmi_category', y='charges', hue='smoker', ax=ax, palette='Set1')
ax.set_title('Charges by BMI Category and Smoking Status')
ax.set_ylabel('Insurance Charges ($)')
ax.set_xlabel('BMI Category')
plt.savefig('bmi_smoker_charges.png', dpi=300, bbox_inches='tight')
print("✅ Saved: bmi_smoker_charges.png")
plt.show()
```

```python
# ============================================================================
# 6. CORRELATION ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("6. CORRELATION ANALYSIS")
print("="*80)

# Create encoded version for correlation
df_encoded = df.copy()
df_encoded['sex'] = df_encoded['sex'].map({'male': 1, 'female': 0})
df_encoded['smoker'] = df_encoded['smoker'].map({'yes': 1, 'no': 0})
df_encoded = pd.get_dummies(df_encoded, columns=['region'], drop_first=True)
df_encoded = df_encoded.drop('bmi_category', axis=1)
```

```python
# Correlation matrix
correlation_matrix = df_encoded.corr()

print("\n🔗 Correlation with Charges:")
print(correlation_matrix['charges'].sort_values(ascending=False))

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, linewidths=1)
plt.title('Correlation Matrix of Features')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
print("✅ Saved: correlation_matrix.png")
plt.show()
```

```python
# ============================================================================
# 7. KEY INSIGHTS & STATISTICS
# ============================================================================

print("\n" + "="*80)
print("7. KEY INSIGHTS & STATISTICS")
print("="*80)

print("\n💡 SMOKER IMPACT:")
smoker_stats = df.groupby('smoker')['charges'].agg(['mean', 'median', 'std'])
print(smoker_stats)
print(f"\nSmokers pay {(smoker_stats.loc['yes', 'mean'] / smoker_stats.loc['no', 'mean'] - 1) * 100:.1f}% more on average")

print("\n💡 AGE IMPACT:")
age_groups = pd.cut(df['age'], bins=[0, 30, 40, 50, 100], labels=['<30', '30-40', '40-50', '50+'])
age_stats = df.groupby(age_groups)['charges'].agg(['mean', 'median', 'count'])
print(age_stats)

print("\n💡 BMI IMPACT:")
bmi_stats = df.groupby('bmi_category')['charges'].agg(['mean', 'median', 'count'])
print(bmi_stats)

print("\n💡 REGION IMPACT:")
region_stats = df.groupby('region')['charges'].agg(['mean', 'median', 'count'])
print(region_stats)

print("\n💡 SEX IMPACT:")
sex_stats = df.groupby('sex')['charges'].agg(['mean', 'median'])
print(sex_stats)

print("\n💡 CHILDREN IMPACT:")
children_stats = df.groupby('children')['charges'].agg(['mean', 'median', 'count'])
print(children_stats)
```

```python
# ============================================================================
# 8. OUTLIER DETECTION
# ============================================================================

print("\n" + "="*80)
print("8. OUTLIER DETECTION")
print("="*80)

# Using IQR method for charges
Q1 = df['charges'].quantile(0.25)
Q3 = df['charges'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['charges'] < lower_bound) | (df['charges'] > upper_bound)]
print(f"\n📊 Number of outliers in charges (IQR method): {len(outliers)}")
print(f"Percentage: {len(outliers)/len(df)*100:.2f}%")
```

```python
# Visualize outliers
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].boxplot(df['charges'], vert=True)
axes[0].set_ylabel('Insurance Charges ($)')
axes[0].set_title('Box Plot of Charges')
axes[0].grid(True, alpha=0.3)

axes[1].scatter(range(len(df)), df['charges'], alpha=0.5, color='blue')
axes[1].axhline(y=upper_bound, color='red', linestyle='--', label='Upper Bound')
axes[1].axhline(y=lower_bound, color='red', linestyle='--', label='Lower Bound')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Insurance Charges ($)')
axes[1].set_title('Charges with Outlier Bounds')
axes[1].legend()

plt.tight_layout()
plt.savefig('outliers_detection.png', dpi=300, bbox_inches='tight')
print("✅ Saved: outliers_detection.png")
plt.show()
```
